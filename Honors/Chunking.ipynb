```python
with open("/content/sample.txt", "r") as f:
    text = f.read()
doc = nlp(text)
noun_chunks = list(doc.noun_chunks)
noun_count = len(noun_chunks)
noun_dict = {}
for noun_chunk in noun_chunks:
  noun = noun_chunk.text
  if noun not in noun_dict:
    noun_dict[noun] = 1
  else:
    noun_dict[noun] += 1
for noun, count in noun_dict.items():
  print(f"{noun}: {count}")
```
